```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

weather = pd.read_csv("D:\\paper\\C\\data\\5\\realityvalues.csv", index_col="date")
weather.columns

x = weather["Temperature"]
y = weather["En_Requests (MJ)"]

idx = np.arange(108)
np.random.shuffle(idx)

#??train
train_idx = idx[:80]
val_idx = idx[80:]

x_train, y_train = x[train_idx], y[train_idx]
x_val, y_val = x[val_idx], y[val_idx]

plt.scatter(x_train, y_train, c='b', edgecolors='w', label='Train')
plt.scatter(x_val, y_val, c='r', edgecolors='w', label='Test')
plt.legend()
plt.show()

#اجرای فرآیند تکراری برای کاهش خطا در رگرسیون خطی
b = np.random.randn(1)
w = np.random.randn(1)
print('w0={:.4f} , b0={:.4f}'.format(w.item(), b.item()))

lr = 0.001
n_epochs = 100000

for epoch in range(n_epochs):
    # Model
    yp = b + w * x_train
    # Loss
    error = (y_train - yp)
    loss = (error ** 2).mean()
    # Gradient
    b_grad = -2 * error.mean()
    w_grad = -2 * (x_train * error).mean()
    # Update 
    b = b - lr * b_grad
    w = w - lr * w_grad

print('wf={:.4f} , bf={:.4f}'.format(w.item(), b.item()))
```

```python
yp = b + w * x_val

error = (yp - y_val)
loss = (error ** 2).mean()

print('Loss=', loss)
```
