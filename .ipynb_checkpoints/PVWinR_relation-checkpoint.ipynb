```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
f=pd.read_csv('D:/paper/A/2/f1/f2/f3/f4/journal/solarenergy/review/results/Horizontal.csv')
df=f
f
x = f["Abs_devision result"]
y = f["UDI %"]
```

```python
f = f[['Abs_devision result','UDI %']]
f
idx = np.arange(25)
np.random.shuffle(idx)
```

```python
#?train
train_idx = idx[:20]
val_idx = idx[20:]

x_train, y_train = x[train_idx], y[train_idx]
x_val, y_val = x[val_idx], y[val_idx]

plt.scatter(x_train, y_train, c='b', edgecolors='w', label='Train')
plt.scatter(x_val, y_val, c='r', edgecolors='w', label='Test')
plt.legend()
plt.show()
```

```python
#تعیین مقدار اولیه برای پارامترهای w و b
w = np.random.randn(1)
b = np.random.randn(1)
print('w0={:.4f} , b0={:.4f}'.format(w.item(), b.item()))

#تخمین مقدار yp برای داده ورودی x
yp = b + w * x_train

#محاسبه مقدار خطا با MSE
error = (yp - y_train)
loss = (error ** 2).mean()
print('Loss={:.4f}'.format(loss))

#تعیین مقادیر جدید برای پارامتر w و b با گرادیان کاهشی
b_grad = -2 * error.mean()
w_grad = -2 * (x_train * error).mean()
print('grad_w={:.4f} , grad_b={:.4f}'.format(w_grad, b_grad))

lr = 0.1
b = b - lr * b_grad
w = w - lr * w_grad
print('w_new={:.4f} , b_new={:.4f}'.format(w.item(), b.item()))
```
